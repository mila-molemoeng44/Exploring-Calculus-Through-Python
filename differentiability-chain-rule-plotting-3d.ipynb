```python
import matplotlib.pyplot as plt 
import sympy as sp
from sympy import symbols
from sympy.plotting import plot


x = symbols('x')
graph_1 = plot(100*sp.cos(x), show = False)
graph_2 = plot(x, show = False)
graph_1.append(graph_2[0])
graph_1
graph_1.show()

```

```python
#Finding the derivative of at a point
import sympy as sp
import math as mt

x = sp.symbols('x')
f = 100* sp.cos(x)

#f_deri = sp.diff(f, x)
#point = mt.pi ** 2
#f_deri_at_2pi = f_deri.subs(x, point)

def derivative(point2):
    f_deri_2 = sp.diff(f, x)
    f_deri_at_x = f_deri_2.subs(x, point2)
    print("The value of the slope of the tangent is: ")
    return f_deri_at_x

point = float(input("What point should we evaluate f(x) at?"))
derivative(point)
```

```python
#Chain Rule

x = sp.symbols('x')

u = sp.sin(x)
f = sp.cos(u)

f_derivative = sp.diff(f, x)
point = 67
f_derivative_at_55 = f_derivative.subs(x, point)

point_2 = mt.pi**67
f_derivative_at_pi = f_derivative.subs(x, point_2)

print(f_derivative)
print(round(float(f_derivative_at_55), 7))
print(round(float(f_derivative_at_pi), 7))
```
